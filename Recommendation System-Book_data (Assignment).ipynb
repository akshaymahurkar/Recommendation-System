```python
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv('D:/ExcelR/Assignments/Downloaded/Recommendation System/book.csv')
books=book.iloc[:,1:]
books.head()
```

```python
book=books.rename(columns={'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'})
book.head()
```

```python
book.info()
```

```python
#No of unique user's in the dataset
len(book.user_id.unique())
```

```python
#No of unique book_title in the dataset
len(book.book_title.unique())
```

```python
book_df =book.pivot_table(index='user_id',
                          columns='book_title',
                          values='book_rating').reset_index(drop=True)
book_df.head()
```

```python
book_df.index=book.user_id.unique()
book_df.head()
```

```python
#fill null values to zero
book_df.fillna(0, inplace=True)
```

```python
book_df.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#store the results in data frame
user_sim_df=pd.DataFrame(user_sim)
```

```python
#set the index and columns in user ids
user_sim_df.index=book.user_id.unique()
user_sim_df.columns=book.user_id.unique()
```

```python
user_sim_df.iloc[:10,:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[:10,:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```

```python
#Finding the correlation with different movies
similar_to_book = book_df.corrwith(book_user_rating)
```

```python
corr_book[corr_book['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(10)
```
